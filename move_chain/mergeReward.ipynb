```python
import pandas as pd
```

```python
# Load CSV files into dataframes
after_migrt_df = pd.read_csv('../data/after_migrt_reward.csv')
before_migrt_df = pd.read_csv('../data/before_migrt_reward.csv')

# Retain only 'fnid' and 'reward' columns from before_migrt_df
before_migrt_df = before_migrt_df[['fnid', 'reward']].rename(columns={'reward': 'pre_reward'})

# Merge dataframes on 'fnid', keeping all rows and fill NaN values with appropriate defaults if needed
merged_df = pd.merge(after_migrt_df, before_migrt_df, on='fnid', how='outer', suffixes=('_after', '_before'))
merged_df.fillna(0, inplace=True)
```

```python
# Reorder columns to have 'before_mig_reward' as the second column
cols = merged_df.columns.tolist()
cols.insert(1, cols.pop(cols.index('pre_reward')))
merged_df = merged_df[cols]
merged_df
```

```python
merged_df.to_csv('../data/all_traj_reward.csv',index=False)
```
